```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
train_df = pd.read_csv('./data/filtered_data/filtered_train.tsv', sep='\t')
test_df = pd.read_csv('./data/filtered_data/filtered_test.tsv', sep='\t')
print('shpae of train data: ', train_df.shape)
print('shpae of test data: ', test_df.shape)
```

```python
# Train & Test 연도별 분포 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

# Train 데이터 연도별 분포
sns.histplot(data=train_df, x='month', ax=axes[0], color='skyblue', kde=True)
axes[0].set_title("Train month distribution")
axes[0].set_xlabel("월")

# Test 데이터 연도별 분포
sns.histplot(data=test_df, x='month', ax=axes[1], color='lightcoral', kde=True)
axes[1].set_title("Test month distribution")
axes[1].set_xlabel("월")

plt.tight_layout()
plt.show()
```

```python
# Train & Test 연도별 분포 시각화
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

# Train 데이터 연도별 분포
sns.histplot(data=train_df, x='year', ax=axes[0], color='skyblue', kde=True)
axes[0].set_title("Train year distribution")
axes[0].set_xlabel("년")
axes[0].set_ylabel("건수")

# Test 데이터 연도별 분포
sns.histplot(data=test_df, x='year', ax=axes[1], color='lightcoral', kde=True)
axes[1].set_title("Test year distribution")
axes[1].set_xlabel("년")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(50, 50))  # 가로/세로 크기 조정
sns.barplot(
    data=train_df,
    y='상권명',
    x='총전기사용량', 
    hue='지역명', 
    dodge=False,
    estimator=np.mean,  
    palette="husl"
)

plt.title("월별 평균 총전기사용량")
plt.xlabel("평균 전기사용량")  
plt.ylabel("상권명") 
plt.xticks(size=0.1)
plt.tight_layout()
plt.show()
```

```python
continous_columns = train_df.select_dtypes(include=['float64', 'int64']).columns.to_list()
for col in continous_columns:
    plt.figure(figsize=(10, 5))
    if col not in ['작성월', '지역코드', '지역명', '상권코드', '상권명', 'month', 'year']:
        sns.histplot(data=train_df, x=col, color='black', kde=True)
        plt.title('Train {} distribution'.format(col))
        plt.tight_layout()
        plt.show()
```

```python
continous_columns = test_df.select_dtypes(include=['float64', 'int64']).columns.to_list()
for col in continous_columns:
    plt.figure(figsize=(10, 5))
    if col not in ['index_origin', '작성월', '지역코드', '지역명', '상권코드', '상권명', 'month', 'year']:
        sns.histplot(data=test_df, x=col, color='black', kde=True)
        plt.title('Test {} distribution'.format(col))
        plt.tight_layout()
        plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='month', y='총전기사용량', data=train_df, palette='pastel')
plt.title("월별 총전기사용량 분포 (Boxplot)")
plt.xlabel("월")
plt.ylabel("총전기사용량")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(x="총전기사용량", bins=50, kde=True, color='cornflowerblue', data=train_df)
plt.title("총전기사용량 분포")
plt.xlabel("총전기사용량")
plt.ylabel("빈도수")
plt.tight_layout()
plt.show()
```
